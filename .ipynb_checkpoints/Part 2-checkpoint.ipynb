```python
#Q12. Create a random dataset of 100 rows and 30 columns. All the values are defined between [1,200]. Perform the following operations:
#(i) Replace all the values with NA in the dataset defined between [10, 60]. Print the count of number rows having missing values.
#(ii) Replace all the NA values with the average of the column value. 
#(iii) Find the Pearson correlation among all the columns and plot heat map. Also select those columns having correlation <=0.7.
#(iv) Normalize all the values in the dataset between 0 and 10.
#(v) Replace all the values in the dataset with 1 if value <=0.5 else with 0.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Create a random dataset
np.random.seed(42)
data = np.random.randint(1, 201, size=(100, 30))
df = pd.DataFrame(data)

# Replace values between 10 and 60 with NA
df.replace(to_replace=range(10, 61), value=np.nan, inplace=True)

# Count rows with missing values
rows_with_na = df.isna().any(axis=1).sum()
print(f'Number of rows with missing values: {rows_with_na}')

# Replace NA values with column averages
df.fillna(df.mean(), inplace=True)

# Pearson correlation and heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation Heatmap')
plt.show()

# Select columns with correlation <= 0.7
low_corr_columns = correlation_matrix.columns[(correlation_matrix.abs() <= 0.7).all(axis=1)]
print(f'Columns with correlation <= 0.7: {list(low_corr_columns)}')

# Normalize values between 0 and 10
scaler = MinMaxScaler(feature_range=(0, 10))
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Replace values with 1 if <= 0.5 else with 0
df_binary = df_normalized.applymap(lambda x: 1 if x <= 0.5 else 0)

# Display the first few rows of the binary dataframe
print(df_binary.head())
```

```python
#Q13. Create a random dataset of 500 rows and 10 columns. 
#Columns 1 to 4 are defined between [-10, 10]; 
#Columns 5 to 8 are defined between [10, 20]; 
#Columns 9 to 10 are defined between [-100, 100]. 
#Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric graph using each algorithm.
#(i) K-Mean clustering
#(ii) Hierarchical clustering
```

```python
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score

# Create a random dataset
np.random.seed(42)
data_q13 = np.hstack((
    np.random.uniform(-10, 10, size=(500, 4)),
    np.random.uniform(10, 20, size=(500, 4)),
    np.random.uniform(-100, 100, size=(500, 2))
))
df_q13 = pd.DataFrame(data_q13)

# K-Means Clustering
inertia = []
silhouette_scores = []
K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(df_q13)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df_q13, kmeans.labels_))

plt.figure(figsize=(12, 6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(K, silhouette_scores, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()

# Hierarchical Clustering
linked = linkage(df_q13, 'ward')
plt.figure(figsize=(12, 6))
dendrogram(linked)
plt.title('Dendrogram')
plt.show()
```

```python
#Q14 - Create a random dataset of 600 rows and 15 columns. All the values are defined between [-100,100]. 
#Perform the following operations:
#(i) Plot scatter graph between Column 5 and Column 6.
#(ii) Plot histogram of each column in single graph.
#(iii) Plot the Box plot of each column in single graph.
```

```python
# Create a random dataset
np.random.seed(42)
data_q14 = np.random.uniform(-100, 100, size=(600, 15))
df_q14 = pd.DataFrame(data_q14)

# Scatter plot between Column 5 and Column 6
plt.figure(figsize=(8, 6))
plt.scatter(df_q14[4], df_q14[5])
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.title('Scatter plot between Column 5 and Column 6')
plt.show()

# Histogram of each column
df_q14.hist(bins=20, figsize=(20, 15))
plt.suptitle('Histogram of each column')
plt.show()

# Box plot of each column
plt.figure(figsize=(20, 10))
df_q14.boxplot()
plt.title('Box plot of each column')
plt.show()
```

```python
#Q15. Create a random dataset of 500 rows and 5 columns:
#All the values are defined between [5,10]. 
#Perform the following operations:
#(i) Perform t-Test on each column.
#(ii) Perform Wilcoxon Signed Rank Test on each column.
#(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
```

```python
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums

# Create a random dataset
np.random.seed(42)
data_q15 = np.random.uniform(5, 10, size=(500, 5))
df_q15 = pd.DataFrame(data_q15)

# Perform t-Test on each column
ttest_results = df_q15.apply(lambda x: ttest_1samp(x, popmean=7.5))
print(f't-Test results:\n{ttest_results}\n')

# Perform Wilcoxon Signed Rank Test on each column
wilcoxon_results = df_q15.apply(lambda x: wilcoxon(x - 7.5))
print(f'Wilcoxon Signed Rank Test results:\n{wilcoxon_results}\n')

# Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
ttest_2sample_result = ttest_ind(df_q15[2], df_q15[3])
wilcoxon_ranksum_result = ranksums(df_q15[2], df_q15[3])
print(f'Two Sample t-Test result between Column 3 and Column 4:\n{ttest_2sample_result}\n')
print(f'Wilcoxon Rank Sum Test result between Column 3 and Column 4:\n{wilcoxon_ranksum_result}\n')
```
